```python
import pandas as pd
import numpy as np
```

```python
read = pd.read_excel('Dry_Bean_Dataset.xlsx')
data = {}
read.sort_values(['Class'], inplace=True)
read.reset_index(inplace=True)
```

```python
read.isnull().values.any()
read.drop(columns=['index'], inplace=True)
X = list(read['Class'].unique())
```

```python
feature = read.drop(columns=['Class'])
feature_names = list(feature.columns)
data['feature_names'] = feature_names
```

```python
classes = list(read['Class'].unique())

data['target_names'] = np.array(classes)
target = []
for i in list(read['Class']):
    if i is X[0]:
        y = 0
        target.append(y)
    elif i is X[1]:
        y = 1
        target.append(y)
    elif i is X[2]:
        y = 2
        target.append(y)
    elif i is X[3]:
        y = 3
        target.append(y)
    elif i is X[4]:
        y = 4
        target.append(y)
    elif i is X[5]:
        y = 5
        target.append(y)
    elif i is X[6]:
        y = 6
        target.append(y)
data['target'] = np.array(target)
```

```python
arr = np.array(feature)
data['data'] = arr
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], random_state=42)
scaler = MinMaxScaler()
scaler.fit(X_train)
x_train = scaler.transform(X_train)
x_test = scaler.transform(X_test)
```

```python
algo = [SVC, GradientBoostingClassifier, RandomForestClassifier]
for i in algo:
    if i is SVC:
        classifier = i()
        classifier.fit(x_train, y_train)
        print(f'{i.__name__} train score : {classifier.score(x_train, y_train)}')
        print(f'{i.__name__} test score : {classifier.score(x_test, y_test)}')
    else:
        classifier = i(random_state=0)
        classifier.fit(x_train, y_train)
        print(f'{i.__name__} train score : {classifier.score(x_train, y_train)}')
        print(f'{i.__name__} test score : {classifier.score(x_test, y_test)}')
```
