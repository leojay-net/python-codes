```python
import pandas as pd
def proportion_of_education():
    # your code goes here
    # YOUR CODE HERE
    read = pd.read_csv('assets/NISPUF17.csv')
    read = read.sort_values('EDUC1')
    a = len(read[read['EDUC1'] == 1]['SEQNUMC'])/len(read['SEQNUMC'])
    b = len(read[read['EDUC1'] == 2]['SEQNUMC'])/len(read['SEQNUMC'])
    c = len(read[read['EDUC1'] == 3]['SEQNUMC'])/len(read['SEQNUMC'])
    d = len(read[read['EDUC1'] == 4]['SEQNUMC'])/len(read['SEQNUMC'])
    return {"less than high school":a,
    "high school":b,
    "more than high school but not college":c,
    "college":d}
proportion_of_education()

def average_influenza_doses():
    # YOUR CODE HERE
    read = pd.read_csv('assets/NISPUF17.csv')
    filtered = read.filter(items=['CBF_01', 'P_NUMFLU']).dropna()
    filtered_01 = filtered.sort_values('CBF_01')
    a = filtered_01[filtered_01['CBF_01'] == 1]
    b = filtered_01[filtered_01['CBF_01'] == 2]
    c = (a['P_NUMFLU'].mean(), b['P_NUMFLU'].mean())
    return c
average_influenza_doses()

def chickenpox_by_sex():
    # YOUR CODE HERE
    read = pd.read_csv('assets/NISPUF17.csv')
    filtered = read.filter(items=['HAD_CPOX', 'SEX','P_NUMVRC']).dropna()
    filtered_01 = filtered.sort_values('SEX')
    a = len(filtered_01[(filtered_01['P_NUMVRC'] >= 1) & (filtered_01['HAD_CPOX'] == 1) & (filtered_01['SEX'] == 1)])
    b = len(filtered_01[(filtered_01['P_NUMVRC'] >= 1) & (filtered_01['HAD_CPOX'] == 2) & (filtered_01['SEX'] == 1)])
    c = len(filtered_01[(filtered_01['P_NUMVRC'] >= 1) & (filtered_01['HAD_CPOX'] == 1) & (filtered_01['SEX'] == 2)])
    d = len(filtered_01[(filtered_01['P_NUMVRC'] >= 1) & (filtered_01['HAD_CPOX'] == 2) & (filtered_01['SEX'] == 2)])
    
    
    return {
        'male':a/b,
        'female':c/d
    }
chickenpox_by_sex()

def corr_chickenpox():
    import scipy.stats as stats
    import numpy as np
    import pandas as pd
    read = pd.read_csv('assets/NISPUF17.csv')
    filtered = read.filter(items=['HAD_CPOX','P_NUMVRC'])
    filtered_01 = filtered[filtered['HAD_CPOX'].isin([1,2])]
    df = filtered_01.dropna()
    #df=pd.DataFrame({"had_chickenpox_column":np.random.randint(1,3,size=(100)),
    #               "num_chickenpox_vaccine_column":np.random.randint(0,6,size=(100))})
    # here is some stub code to actually run the correlation
    corr, pval=stats.pearsonr(df["HAD_CPOX"],df["P_NUMVRC"])
    
    # just return the correlation
    return corr
corr_chickenpox()
```
